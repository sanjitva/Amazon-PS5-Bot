## Final Code (Google Chrome)

```python
def amazon_bot_chrome():
    
    from selenium import webdriver
    from selenium.webdriver.common.keys import Keys
    import chromedriver_autoinstaller
    chromedriver_autoinstaller.install()

    URL1 = 'https://www.amazon.com/PlayStation-5-Console/dp/B08FC5L3RG'
    URL2 = 'https://www.amazon.com/PlayStation-5-Digital/dp/B08FC6MR62'

    WAIT_TIME = 5

    username = "sgottum1@binghamton.edu"
    password = "masterviva"


    #Open the browser with the link
    driver = webdriver.Chrome()
    driver.get(URL1)

    #Click on Login button
    login_elem = driver.find_element_by_id("nav-link-accountList")
    login_elem.click()

    #Enter Username
    username_elem = driver.find_element_by_xpath("//input[@name='email']")
    username_elem.clear()
    username_elem.send_keys(username)
    driver.implicitly_wait(10)
    username_elem.send_keys(Keys.RETURN) #Presses the Enter Key on your keyboard

    #Enter Password
    driver.implicitly_wait(10)
    password_elem = driver.find_element_by_xpath("//input[@name='password']")
    password_elem.clear()
    password_elem.send_keys(password)
    password_elem.send_keys(Keys.RETURN) #Presses the Enter Key on your keyboard


    #Refresh page until Buy Now Option is available


    availability = driver.find_element_by_class_name('a-color-price').text

    while availability == 'Currently unavailable.':
        driver.refresh()
        
    else:
        #Click on Buy Now Button
        driver.implicitly_wait(10)
        buy_now = driver.find_element_by_id('submit.buy-now')
        buy_now.click()

        try:
            #Click on Purchase Now Button if Popup page appears
            driver.implicitly_wait(10)
            iframe = driver.find_element_by_id("turbo-checkout-iframe")
            driver.switch_to.frame(iframe)
            driver.implicitly_wait(10)
            PurchaseNow_button = driver.find_element_by_class_name("a-button-inner")
            PurchaseNow_button.click()
            driver.implicitly_wait(10)

        except:
            #Click on Place Order Button if directed to Checkout Page
            driver.implicitly_wait(10)
            place_order = driver.find_element_by_name('placeYourOrder1')
            place_order.click()
            driver.implicitly_wait(10)
    
    print(f'Purchase Successful, Price of Product: ${availability}')
```
